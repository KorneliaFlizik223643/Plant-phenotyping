```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

#### Note: Task 2 was used to crop the images. 

```python
def  instance_segmentation(image_path):
    image = cv2.imread(image_path)
    # Convert the image to a grayscale
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Convert the image to a binary
    th, image_binary = cv2.threshold(grayscale, 161, 202, cv2.THRESH_BINARY_INV)
    # Extract connected components using connectedComponentsWithStats
    retval, labels, stats, centroids = cv2.connectedComponentsWithStats(image_binary, 8, cv2.CV_32S)
    # Creating a blank image
    blank = np.zeros_like(image_binary)
    # Filtering elements based on criteria
    for label, stat in enumerate(stats):
        x, y, w, h, a = stat
        if h> 300 and a > 3000:
            # Drawing the elements on the blank image
            mask = (labels == label).astype(np.uint8)
            blank[mask > 0] = label+1

    print(plt.imshow(blank))
```
